```python
from PIL import Image
import os

def getfilenames(mulu):  
    filenames=[]
    for filename in os.listdir(mulu):
        filenames.append(filename)
    return filenames

def compress(dir, filename):
    fig_name = os.path.join(dir, filename)

    try:
        img=Image.open(fig_name)
        save_args = {'format':img.format}
        save_args['optimize']=True
        save_args['quality']=25
        save_args['progressive=True']=True

        img.save(os.path.join(dir, '{}'.format(filename)), **save_args)
    except IOError as e:
        print(e)
```

```python

dir = '/mnt/d/BaiduNetdiskDownload/Figures'   
file_names = getfilenames(dir)

for filename in file_names:
    compress(dir, filename)
    print(dir, filename)
```
